Number of samples needed to get accurate coverage estimation. If 95% CI is used. $\sigma$ is the standard error.

$\sigma^2 = \text{Var}(C) = \frac{p(1-p)}{n} = \frac{0.95*0.05}{n}$
So, $n = \frac{0.95*0.05}{\sigma^2}$

```python
sigma_coverage_error = 0.005
number_samples = 0.95 * 0.05 / sigma_coverage_error**2
print(number_samples)

sigma_coverage_error = 0.01
number_samples = 0.95 * 0.05 / sigma_coverage_error**2
print(number_samples)

n = 500
sigma_hat = np.sqrt(0.95 * 0.05 / n)
print(sigma_hat)
```

KS test CI bandwidth.
$\frac{ c(\alpha) }{\sqrt{\text{# samples}}}, \quad 
c(\alpha) =  \sqrt{-\log\left(\frac{\alpha}{2}\right)\frac{1}{2}}$

```python
test_size = 0.001
c_a = np.sqrt(-np.log(test_size/2)/2)
print(c_a)
```

#### CI for pearson correlation.
This will be the lower bound for the CI coverage.

```python
def corr_se(r,n):
    z_prime = 0.5 * np.log((1 + r) / (1 - r))
    se = 1 / np.sqrt(n - 3) # Sample standard error
    CI_lower, CI_upper = z_prime - se, z_prime + se
    left_se, right_se = np.tanh(z_prime)-np.tanh(CI_lower), np.tanh(CI_upper)-np.tanh(z_prime)
    return np.array([left_se, right_se])
rhos = np.arange(0, 0.9, 0.001); ses = np.zeros([len(rhos), 2])
for i, rho in enumerate(rhos):
    ses[i] = corr_se(rho, 15*13)
plt.figure(figsize=[5, 2])
plt.plot(rhos, ses)
plt.grid(); plt.ylim(0); plt.show()
print(np.max(ses), np.min(ses))
```

## Verify proposal distribution scale. This for update_q_arc

```python
Sigma = model.sigma_cross_pop
# proposal_cov = Sigma * 0.02  # True cov.
# proposal_cov = np.diag(np.diag(Sigma)) * 0.01  # diag cov.
noise = np.random.randn(Sigma.shape[0], Sigma.shape[1]) * 0.005  # True + noise
proposal_cov = (Sigma + noise.T@noise) * 0.02

mu = np.zeros(Sigma.shape[0])
q_prev = np.ones(Sigma.shape[0]) * 0.1
num_samples = 3000
samples = np.zeros([num_samples, Sigma.shape[0]])
log_likelihood_list = np.zeros(num_samples)
accept_cnt = 0

for i in range(num_samples):
    q_cnd = np.random.multivariate_normal(q_prev, proposal_cov)
    log_likelihood_prev = scipy.stats.multivariate_normal.logpdf(q_prev, mu, Sigma)
    log_likelihood_cnd = scipy.stats.multivariate_normal.logpdf(q_cnd, mu, Sigma)
    mh_ratio = np.exp(log_likelihood_cnd - log_likelihood_prev)
    log_likelihood_list[i] = log_likelihood_cnd
    if np.random.rand() < mh_ratio:
        q_prev = q_cnd
        accept_cnt += 1
    samples[i] = q_prev
```

```python
print('Accept ratio:', accept_cnt / num_samples)
plt.figure(figsize=[10, 3])
plt.plot(log_likelihood_list)
plt.show()
```
