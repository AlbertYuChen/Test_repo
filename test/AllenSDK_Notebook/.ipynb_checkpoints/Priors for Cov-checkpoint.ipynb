### Inverse Wishart.

```python
num_samples = 5000
p = 9
nu = p + 1
scale = np.diag([1] * p) * 4  # Same diag values.
```

```python
# scale = np.diag(np.random.rand(6) * 5)  # Different diag vales.
S = np.zeros([num_samples, p, p])
for i in range(num_samples):
    S[i] = scipy.stats.invwishart.rvs(scale=scale, df=nu)
```

```python
mean = np.median(S, axis=0)
# print(S.shape)
# print(mean)
```

```python
corr0 = S[:,0,1] / np.sqrt(S[:,0,0] * S[:,1,1])
corr1 = S[:,0,2] / np.sqrt(S[:,0,0] * S[:,2,2])
corr2 = S[:,1,2] / np.sqrt(S[:,1,1] * S[:,2,2])
# corr5 = S[:,0,5] / np.sqrt(S[:,0,0] * S[:,5,5])

Omega = np.linalg.inv(S)
pcorr0 = -Omega[:,0,1] / np.sqrt(Omega[:,0,0] * Omega[:,1,1])
pcorr1 = -Omega[:,0,2] / np.sqrt(Omega[:,0,0] * Omega[:,2,2])
pcorr2 = -Omega[:,1,2] / np.sqrt(Omega[:,1,1] * Omega[:,2,2])
```

```python
x = np.linspace(0, 1.0, 200)
x_plot = 2 * x - 1
# y = scipy.stats.beta.pdf(x, 1.5, 1.5) / 2
y = scipy.stats.beta.pdf(x, 1, 1) / 2

plt.figure(figsize=[12,3])
plt.subplot(131)
seaborn.distplot(corr0, kde=False, norm_hist=True)
plt.plot(x_plot, y, 'r')
plt.subplot(132)
seaborn.distplot(corr1, kde=False, norm_hist=True)
plt.subplot(133)
seaborn.distplot(corr2, kde=False, norm_hist=True)
plt.title(f'inv-Wishart dim={p}')
plt.show()
```

```python
x = np.linspace(0, 1.0, 200)
x_plot = 2 * x - 1
beta = (nu-1)/2
y = scipy.stats.beta.pdf(x, beta, beta) / 2
plt.figure(figsize=[12,3])
plt.subplot(131)
seaborn.distplot(pcorr0, kde=False, norm_hist=True)
plt.plot(x_plot, y, 'r')
# plt.subplot(132)
# seaborn.distplot(pcorr1, kde=False, norm_hist=True)
# plt.subplot(133)
# seaborn.distplot(pcorr2, kde=False, norm_hist=True)
plt.title(fr'IW  $d$={p}  $\nu$={nu}')
plt.show()
```

```python
r0 = scipy.stats.pearsonr(np.abs(corr0), np.abs(corr1)); print(r0)
r = scipy.stats.pearsonr(np.abs(corr0), np.abs(corr2)); print(r)
r = scipy.stats.pearsonr(np.abs(corr1), np.abs(corr2)); print(r)

plt.figure(figsize=[3, 3])
plt.plot(corr0, corr1, 'k.', ms=1)
plt.axis('equal')
plt.title(fr'IW  $d$={p}  $\nu$={nu}  $r$={r[0]:.2f}', fontsize=14)
plt.xlabel(r'$\rho_{1,2}$', fontsize=14)
plt.ylabel(r'$\rho_{1,3}$', fontsize=14)
plt.xlim(-1.1, 1.1)
plt.ylim(-1.1, 1.1)
plt.xticks([-1, 1])
plt.yticks([-1, 1])

figure_path = f'D:/Brain_Network/Output/fig/IW_dim{p}_df{nu}_distribution.pdf'
plt.savefig(figure_path, bbox_inches='tight')

plt.show()
```

```python
x = np.linspace(0, 10.0, 1000)
p = 5
nu = p + 1
alpha = (nu - p + 1) / 2
beta = 2
y = scipy.stats.invgamma.pdf(x, a=alpha, scale=beta)
mode = beta / (alpha + 1)
quantile_mode = scipy.stats.invgamma.cdf(mode, a=alpha, scale=beta)
quantile_beta = scipy.stats.invgamma.cdf(beta, a=alpha, scale=beta)
quantiles = [quantile_mode, 0.05, 0.5, 0.7, 0.9, 0.95]
q_vals = scipy.stats.invgamma.ppf(quantiles, a=alpha, scale=beta)

print('beta quantile', quantile_beta)
print('mode quantile', q_vals[0])
print(quantiles)
print(q_vals)

plt.figure(figsize=[15, 6])
plt.subplot(221)
# seaborn.distplot(S[:,0,0], bins=np.linspace(0, 100, 1000) , kde=False, norm_hist=True)
plt.plot(x, y, 'k')
plt.xlim(0, 8)
for q_val in q_vals:
    plt.axvline(x=q_val, c='k', ls='--')
plt.ylim(bottom=0)

plt.text(q_vals[1]-0.73, 0.3, f'{quantiles[1]*100:.1f}%', fontsize=12)
plt.text(mode-0.2, 0.3, f'{quantile_mode*100:.1f}%', fontsize=12)
plt.text(beta, 0.3, f'{quantile_beta*100:.1f}%', fontsize=12)
plt.text(q_vals[2], 0.3, f'{quantiles[2]*100:.1f}%', fontsize=12)
plt.text(q_vals[3], 0.3, f'{quantiles[3]*100:.1f}%', fontsize=12)

figure_path = 'D:/Brain_Network/Output/fig/inv_gamma_eg.pdf'
plt.savefig(figure_path, bbox_inches='tight')
plt.show()
```

### LKJ prior

```python
import pymc3 as pm
p = 3
num_samples = 100000
# eta_list = [0.1, 0.5, 1, 2, 10]
eta_list = [1]
if p == 2:
    z_eta = np.zeros([len(eta_list), num_samples])
elif p > 2:
    z_eta = np.zeros([len(eta_list), num_samples, p*(p-1)//2])
with pm.Model() as model:
    for i, eta in enumerate(eta_list):
        cov = pm.LKJCorr(f'cov{i}', n=p, eta=eta)
        z_eta[i] = cov.random(size=num_samples)
```

```python
alpha_k = eta + 0.5 * p - 1
print(f'alpha_k:{alpha_k}')

for i, eta in enumerate(eta_list):
    # Theoretical distribution.
    x = np.linspace(0, 1.0, 200)
    x_plot = 2 * x - 1
    y = scipy.stats.beta.pdf(x, alpha_k, alpha_k) / 2

    z = z_eta[i]   
    plt.figure(figsize=[12,3])
    plt.subplot(131)
#     CI_left = np.quantile(z[:,0], 0.025)
#     CI_right = np.quantile(z[:,0], 0.975)
    data = z[:,0] if p > 2 else z
    seaborn.distplot(data, kde=False, norm_hist=True)
    plt.plot(x_plot, y, 'r')
#     plt.axvline(x=CI_left, color='k')
#     plt.axvline(x=CI_right, color='k')
#     plt.title(f'eta={eta}')
    plt.title(f'flat dim={p}')
#     plt.subplot(132)
#     seaborn.distplot(z[:,1])
#     plt.subplot(133)
#     seaborn.distplot(z[:,2])
    plt.show()
```
