# Get the covariance matrix

```python
orientation = 45  # 'null'
# temporal_frequency = 8.  # 'null'

stimulus_table = session.get_stimulus_table(['drifting_gratings'])
# trials_indices = stimulus_table[
#     (stimulus_table['orientation']==orientation) & 
#     (stimulus_table['temporal_frequency']==temporal_frequency)].index.values
trials_indices = stimulus_table[
    (stimulus_table['orientation']==orientation)].index.values

print('num trials:', len(trials_indices))

# display(session.units.head())
units_indices = session.units.index.values
print('num units:', len(units_indices))
# print(selected_units['probe_description'].values)
```

### Select neurons by region

```python
# util.VISUAL_AREA  util.THALAMUS_AREA
selected_units = session.units[
    session.units['ecephys_structure_acronym'].isin(util.VISUAL_AREA)]

display(selected_units.head())
display(selected_units.shape)
# display(selected_units['ecephys_structure_acronym'].values)
```

```python
model = data_model.AllenInstituteDataModel(session)
spike_counts = model.get_trial_metric_per_unit_per_trial(
        stimulus_presentation_ids=trials_indices,
        unit_ids=selected_units.index.values,
        trial_time_window=[0.0, 2.2],
        metric_type='count')
display(spike_counts.shape)
# display(spike_counts.head())
```

```python
corr_matrix = spike_counts.T.corr()
visualizer = data_visualizer.AllenInstituteDataVisualizer(session)
visualizer.plot_matrix_by_areas(
    corr_matrix,
    fig_title='test')
```

```python
corr_matrix = spike_counts.T.corr()
corr_matrix[corr_matrix < 0.2] = 0

visualizer = data_visualizer.AllenInstituteDataVisualizer(session)
visualizer.plot_matrix_by_areas(
    corr_matrix,
    fig_title='test')
```

## Separate the spike counts correlation matrix

```python
corr_matrix = spike_counts.T.corr()
model = data_model.AllenInstituteDataModel(session)
sub_group_df = model.seperate_probe_pairs_units_by_corr_matrix(corr_matrix, corr_threshold=0.35)
probe0 = 'probeC'
probe1 = 'probeE'

sub_units0 = sub_group_df[(sub_group_df['probe_from'] == probe0) & 
                          (sub_group_df['probe_to'] == probe1) & 
                          (sub_group_df['included'] == True)]['unit_ids'].values
sub_units1 = sub_group_df[(sub_group_df['probe_from'] == probe1) & 
                          (sub_group_df['probe_to'] == probe0) & 
                          (sub_group_df['included'] == True)]['unit_ids'].values
sub_units0c = sub_group_df[(sub_group_df['probe_from'] == probe0) & 
                           (sub_group_df['probe_to'] == probe1) & 
                           (sub_group_df['included'] == False)]['unit_ids'].values
sub_units1c = sub_group_df[(sub_group_df['probe_from'] == probe1) & 
                           (sub_group_df['probe_to'] == probe0) & 
                           (sub_group_df['included'] == False)]['unit_ids'].values

units = session._filter_owned_df('units', ids=corr_matrix.index.values)
probe_units0 = units[units['probe_description'] == probe0].index.values
probe_units1 = units[units['probe_description'] == probe1].index.values

print(sub_units0)
print(sub_units1)
print(sub_units0c)
print(sub_units1c)
print(probe_units0)
```

```python
sub_index0 = probe_units0
sub_index1 = probe_units1

# sub_index0 = sub_units0
# sub_index1 = sub_units1

# sub_index0 = np.hstack((sub_units0, sub_units0c))
# sub_index1 = np.hstack((sub_units1, sub_units1c))

sub_corr_matrix = corr_matrix.loc[sub_index0, sub_index1]
visualizer = data_visualizer.AllenInstituteDataVisualizer(session)
visualizer.plot_matrix_by_areas(
    sub_corr_matrix,
    fig_title='test')
```
