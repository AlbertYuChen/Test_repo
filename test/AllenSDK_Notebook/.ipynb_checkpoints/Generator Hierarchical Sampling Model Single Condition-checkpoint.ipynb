# Simulation Test

```python
generator_model = hierarchical_model_generator.HierarchicalModelGenerator()
output_folder = 'D:/Brain_Network/test/'
# output_folder = '/Users/chenyu/Workspace/Brain_Network/test/'
file_path = (output_folder + 
    'hierarchical_bayesian_model_pseudo_spikes_B_R15_0_500ms_20200629-165401.pkl')
generator_model.load_data(file_path)

trial_time_window=generator_model.trial_time_window
spike_train_time_line = generator_model.spike_train_time_line
dt=generator_model.dt
spike_trains, spike_times = generator_model.spike_trains, generator_model.spike_times
session = None
spike_counts, spike_shifts = generator_model.spike_counts, generator_model.spike_shifts
units_probes = generator_model.selected_units['probe_description']
probes = generator_model.probes
selected_units = generator_model.selected_units
trials_indices = generator_model.trials_indices
```

```python
model.plot_group_intensity_per_trial(spike_trains, spike_train_time_line, trials_indices, probes)
```

# Initialization

```python
model = hierarchical_sampling_model_single_condition.HierarchicalSamplingModel(session)
np.random.seed(0)
model.initial_step(spike_trains, spike_times, spike_train_time_line,
                   selected_units, trials_indices, trial_time_window, probes,
                   model_feature_type='B', eta_smooth_tuning=1e-8, verbose=False)
model.group_activity_statistic(probes, trials_indices, spike_train_time_line)
```

```python
model.group_activity_statistic(probes, trials_indices, spike_train_time_line)
```

# Full loop

```python
t = trange(0, 200)
for itr in t:
    model.update_f_local_pop_gac(spike_trains, spike_train_time_line, probes, trials_indices,
                                 verbose=False)
    model.update_f_cross_pop_ac(spike_trains, spike_times, spike_train_time_line, probes, trials_indices,
                                separate_fitting=True, verbose=False)
#     model.update_f_cross_local_pop_ac(spike_trains, spike_train_time_line, probes, trials_indices, spike_times,
#                                       verbose=False)
    if itr < 20:
        model.update_q_arc(spike_trains, spike_train_time_line, probes, trials_indices,
                           sample_type='fit', fit_peak=True, verbose=False)
    if itr >= 20:
        model.update_q_arc(spike_trains, spike_train_time_line, probes, trials_indices,
                           sample_type='sample', proposal_scalar=0.15, verbose=False)
    if itr >= 30:
        model.update_mu_simga(sample_type='sample', verbose=False)

    if itr > 10:
        model.update_z_ngac(spike_trains, spike_train_time_line, probes, trials_indices, 
                            sample_type='sample', verbose=False)
        model.update_p_gac(probes, sample_type='sample', verbose=False)

    model.complete_log_likelihood(spike_trains, spike_train_time_line, trials_indices, probes)
    t.set_description(f'log-likelihood: {model.samples.log_likelihood[-1]:.2f}')
```
