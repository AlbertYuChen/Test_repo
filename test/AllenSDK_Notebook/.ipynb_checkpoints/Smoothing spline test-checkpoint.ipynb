### trial-wise offsets.

constant trial-wise offset. non-constant fit.

```python
import smoothing_spline
fit_model = smoothing_spline.SmoothingSpline()

time_line = np.linspace(0, 2, int(1.995 / 0.005))
log_lmbd_0 = fit_model.create_sine_wave(
    time_line, frequency=4, show_figure=False)
log_lmbd_0[180:190] = 1.8
log_lmbd_1 = log_lmbd_0 - 3
log_lmbd_2 = log_lmbd_0 - 1

spikes1 = fit_model.generate_spike_trains(np.exp(log_lmbd_1), 20, random_seed=623)
spikes2 = fit_model.generate_spike_trains(np.exp(log_lmbd_2), 20, random_seed=623)
spikes = np.vstack([spikes1, spikes2])

log_lambda_offset = np.zeros((40 , 1))
log_lambda_offset[:20, 0] = -3
log_lambda_offset[20:, 0] = -1

log_lmbda_hat, beta = fit_model.poisson_regression_smoothing_spline(
    spikes, time_line,
    log_lambda_offset=log_lambda_offset, 
    lambda_tuning=1e-7, max_num_iterations=2000, verbose=True)

plt.figure()
plt.plot(log_lmbda_hat, label='Estimated')
plt.plot(log_lmbd_0, 'r:', label='True 1')
plt.legend()
plt.ylim(-3,3)
plt.show()
```

trial-wise non-constant offset, non-constant fit

```python
import smoothing_spline
fit_model = smoothing_spline.SmoothingSpline()

time_line = np.linspace(0, 2, int(1.995 / 0.005))
log_lmbd_0 = fit_model.create_sine_wave(
    time_line, frequency=4, show_figure=False)
log_lmbd_0[180:190] = 1.8
log_lmbd_1 = log_lmbd_0 - 3
log_lmbd_2 = log_lmbd_0 - 1

spikes1 = fit_model.generate_spike_trains(np.exp(log_lmbd_1), 20, random_seed=623)
spikes2 = fit_model.generate_spike_trains(np.exp(log_lmbd_2), 20, random_seed=623)
spikes = np.vstack([spikes1, spikes2])

log_lambda_offset = np.zeros((40 , len(time_line)))
log_lambda_offset[:20, :] = log_lmbd_0 - 3
log_lambda_offset[20:, :] = log_lmbd_0 - 1

log_lmbda_hat, beta = fit_model.poisson_regression_smoothing_spline(
    spikes, time_line,
    log_lambda_offset=log_lambda_offset, 
    lambda_tuning=1e-7, max_num_iterations=2000, verbose=True)

plt.figure()
plt.plot(log_lmbda_hat + log_lmbd_0, '.-', label='log-lambda with offset')
plt.plot(log_lmbda_hat, label='log-lambda without offset')
plt.plot(log_lmbd_0, 'g', label='True 1')
plt.legend()
plt.ylim(-3,3)
plt.show()
```

```python
import smoothing_spline
fit_model = smoothing_spline.SmoothingSpline()

# num_samples = 400
# lmbd = -4
# log_lmbd = lmbd * np.ones(num_samples)
# spikes = fit_model.generate_spike_trains(
#     np.exp(log_lmbd), 10, random_seed=623)

time_line = np.linspace(0, 2, int(1.995 / 0.005))
log_lmbd = fit_model.create_sine_wave(
    time_line, frequency=4, show_figure=False)
log_lmbd[180:190] = 1.8
log_lmbd += -1.9
spikes  = fit_model.generate_spike_trains(np.exp(log_lmbd), 40, random_seed=623)

lambda_tuning_list = 10 ** np.arange(-10, -5, 0.5)
print(np.arange(-5, 3, 0.5))
print(lambda_tuning_list)

nll_test_array = fit_model.poisson_regression_smoothing_spline_CV(
    spikes, 
    time_line, 
    lambda_tuning_list=lambda_tuning_list, 
    max_num_iterations=2000,
    verbose=False)


plt.figure()
plt.plot(lambda_tuning_list, nll_test_array.mean(axis=1))
plt.xscale('log', basex=10) 
plt.show()
```

```python

lambda_optimal_idx = np.argmin(nll_test_array.mean(axis=1))
lambda_optimal = lambda_tuning_list[lambda_optimal_idx]
print(lambda_optimal)


log_lmbda_hat, beta = fit_model.poisson_regression_smoothing_spline(
    spikes, time_line, 
    lambda_tuning=lambda_optimal, 
    max_num_iterations=1000)

plt.figure()
plt.plot(log_lmbda_hat, label='Estimated')
plt.plot(log_lmbd, label='True')
plt.legend()
plt.ylim(-5,1)
plt.show()
```
